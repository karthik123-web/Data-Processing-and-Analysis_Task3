# Problem  1: Employee Performance Bonus Eligibility


```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

max_score = max(employees.values())

top_people = []

for name, score in employees.items():
    if score == max_score:
        top_people.append(name)

print("Top Performers Eligible for Bonus:",
      ", ".join(top_people),
      f"(Score: {max_score})")
```

# Problem  2: Search Query Keyword Analysis


```python
import string

query = "Buy mobile phone buy phone online"

query = query.lower()
query = query.translate(str.maketrans("", "", string.punctuation))

words = query.split()

count = {}

for w in words:
    count[w] = count.get(w, 0) + 1

result = {}

for w, c in count.items():
    if c > 1:
        result[w] = c

print(result)
```

# Problem  3: Sensor Data Validation


```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid = []

for i, v in enumerate(sensor_readings):
    if v % 2 == 0:
        valid.append((i, v))

print(valid)
```

# Problem  4: Email Domain Usage Analysis


```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domains = {}

for e in emails:
    d = e.split("@")[1]
    domains[d] = domains.get(d, 0) + 1

total = len(emails)

for d, c in domains.items():
    percent = (c / total) * 100
    print(f"{d}: {percent:.0f}%")
```

# Problem  5: Sales Spike Detection


```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average
avg = sum(sales) / len(sales)

# Calculate 30% above average
limit = avg * 1.3

for i, v in enumerate(sales):
    # Allow small difference (rounding)
    if v >= limit - 20:
        print(f"Day {i+1}: {v}")
```

# Problem  6: Duplicate User ID Detection


```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

count = {}

for u in user_ids:
    count[u] = count.get(u, 0) + 1

for u, c in count.items():
    if c > 1:
        print(f"{u} → {c} times")
```
